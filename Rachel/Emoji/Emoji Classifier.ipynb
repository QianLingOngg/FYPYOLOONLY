```python
# importing libraries
import pandas as pd
import numpy as np
import demoji
import sys
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
# from emosent import get_emoji_sentiment_rank

# datafile
# data_file = "emoji test file.csv"
# data_file = "SentimentAnalysis_v2_labelled_polarity.csv"
data_file = "Week 4/Data/Shopee_Maybelline.csv"
# data_file = open(data_file, errors='ignore')
data = pd.read_csv(data_file)
data.columns = data.columns.str.strip().str.replace(" ","_") # reason for doing this is cause cannot get columns with space into a list
data.info()

data.head()
```

```python
review_list = data['Review'].tolist()
# polarity_list = data['Polarity'].tolist()
new_emoji_list = []
new_review_list = []

for each_review in review_list:
    emoji_dictionary = demoji.findall(each_review)
#     print (emoji_dictionary)
    if len(emoji_dictionary) == 0:
        new_emoji_list.append([''])
    else:
        temp_list = []
        for key in emoji_dictionary:
            temp_list.append(key)
        new_emoji_list.append(temp_list)
        new_review_list.append(each_review)

print (new_emoji_list)
print (new_review_list)
print (len(new_emoji_list))
print (len(new_review_list))
# List within a list (where each review is tag to one list containing the emojis in the review)
# print (len(emoji_list))

# print (type(new_emoji_list[0]))

new_data = {'Review': new_review_list, 'Emoji': new_emoji_list}
new_df = pd.DataFrame.from_dict(new_data)
new_df.to_csv("Emoji Reviews from Maybelline.csv")
```

2. Heart Eye == 0.5 Neutral, 0.5 Positive
3. Red Heart == 0.75 Negative, 0.25 Neutral
4. Tick == 1.0 Neutral

```python
analyzer = SentimentIntensityAnalyzer()

def emoji_scorer(list_of_emoji):
    emoji_string = ""
    for each_emoji in list_of_emoji:
        emoji_string += each_emoji
    score = analyzer.polarity_scores(emoji_string)
    return (emoji_string, score)

for emoji in emoji_list:
    if emoji != "":
        print (emoji_scorer(emoji))
```
