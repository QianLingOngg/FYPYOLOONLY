```python
# importing libraries
import pandas as pd
import numpy as np
import demoji
import sys
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
# from emosent import get_emoji_sentiment_rank

# datafile
# data_file = "emoji test file.csv"
# data_file = "SentimentAnalysis_v2_labelled_polarity.csv"
data_file = "Week 4/Data/Shopee_Maybelline.csv"
# data_file = open(data_file, errors='ignore')
data = pd.read_csv(data_file)
data.columns = data.columns.str.strip().str.replace(" ","_") # reason for doing this is cause cannot get columns with space into a list
data.info()

data.head()
```

```python
review_list = data['Review'].tolist()
# polarity_list = data['Polarity'].tolist()
new_emoji_list = []
new_review_list = []

for each_review in review_list:
    emoji_dictionary = demoji.findall(each_review)
#     print (emoji_dictionary)
    if len(emoji_dictionary) == 0:
        pass
    else:
        temp_list = ""
        for key in emoji_dictionary:
            temp_list += key + " "
        temp_list = temp_list[0:-1]
        new_emoji_list.append(temp_list)
        new_review_list.append(each_review)

print (new_emoji_list)
print (new_review_list)
print (len(new_emoji_list))
print (len(new_review_list))
# List within a list (where each review is tag to one list containing the emojis in the review)
# print (len(emoji_list))

# print (type(new_emoji_list[0]))

new_data = {'Review': new_review_list, 'Emoji': new_emoji_list}
new_df = pd.DataFrame.from_dict(new_data)
new_df.to_csv("Emoji Reviews from Maybelline.csv")
```

## Vader 
However, the polarity are not very accurate given the context of our problem statement.

Example: 
1. Thumbs Up == 1.0 Neutral
2. Heart Eye == 0.5 Neutral, 0.5 Positive
3. Red Heart == 0.75 Negative, 0.25 Neutral
4. Tick == 1.0 Neutral

```python
analyzer = SentimentIntensityAnalyzer()

def emoji_scorer(list_of_emoji):
    emoji_string = ""
    for each_emoji in list_of_emoji:
        emoji_string += each_emoji
    score = analyzer.polarity_scores(emoji_string)
    return (emoji_string, score)

for emoji in new_emoji_list:
    if emoji != "":
        print (emoji_scorer(emoji))
```

## Emoji Sentiment Ranking

Reference:
1. http://kt.ijs.si/data/Emoji_sentiment_ranking/index.html
2. https://pypi.org/project/emosent-py/


```python
data_file = "Emoji Reviews from Maybelline.csv"
# data_file = open(data_file, errors='ignore')
data = pd.read_csv(data_file)
data.columns = data.columns.str.strip().str.replace(" ","_") # reason for doing this is cause cannot get columns with space into a list

data.info()
data.head()

emoji_sentiment_ranking_file = "Emoji_Sentiment_Data_v1.0.csv"
emoji_sentiment_ranking = pd.read_csv(emoji_sentiment_ranking_file)
emoji_sentiment_ranking.columns = emoji_sentiment_ranking.columns.str.strip().str.replace(" ","_")
# emoji_sentiment_ranking.info()
emoji_sentiment_ranking.head()

# data.head()
```

```python
reviews_with_emoji = data['Review'].tolist()
emoji = data['Emoji'].tolist()

# print (emoji)

emoji_unicode = emoji_sentiment_ranking['Unicode_codepoint'].tolist()
occurences = emoji_sentiment_ranking['Occurrences'].tolist()
positive = emoji_sentiment_ranking['Positive'].tolist()
neutral = emoji_sentiment_ranking['Neutral'].tolist()
negative = emoji_sentiment_ranking['Negative'].tolist()

def getSentiment(emoji_string):
#     emoji_string = "ğŸ˜„ â¤ï¸"
    # initialise
    positive_sentiment = 0
    neutral_sentiment = 0
    negative_sentiment = 0
    
    emoji_splitted = emoji_string.split(" ")
    
    for i in range(len(emoji_splitted)):
        current_emoji = emoji_splitted[i]
        current_emoji = current_emoji[0]
        if len(current_emoji) == 1:
            unicode = '0x{:X}'.format(ord(current_emoji)).lower()
            
            if unicode in emoji_unicode:
                find_unicode_index = emoji_unicode.index(unicode)
                positive_sentiment += int(positive[find_unicode_index])
                neutral_sentiment += int(neutral[find_unicode_index])
                negative_sentiment += int(negative[find_unicode_index])
        else: 
            pass
    
#     print (positive_sentiment, "Positive")
#     print (neutral_sentiment, "Neutral")
#     print (negative_sentiment, "Negative")
    
    if positive_sentiment > neutral_sentiment and negative_sentiment:
        return ("Positive")
    elif negative_sentiment > neutral_sentiment and positive_sentiment:
        return ("Negative")
    else:
        return ("Neutral")

for each_emoji in emoji:
    current_review = each_emoji
    print (current_review)
    print (getSentiment(current_review))
    print ("\n")
```
