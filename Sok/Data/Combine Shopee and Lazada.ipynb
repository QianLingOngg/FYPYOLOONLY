```python
import numpy as np
import pandas as pd


shopee_file = "SHOPEE_ALL_BRANDS_13Feb2020.csv"
shopee = pd.read_csv(shopee_file)
shopee["Platform"] = "Shopee"
cols = ["Brand","Category","Product Name ","Price","Reviewer","Review","Product Purchase","Ratings","Date Of Review","Response"]

shopee = shopee[cols]

lazada_file = "Lazada_AllProducts_160220.csv"
lazada = pd.read_csv(lazada_file)
lazada["Platform"] = "Lazada"

lazada_price = lazada["Price"]
lazada_price = [p.strip("$") for p in lazada_price]

lazada["Price"] = lazada_price
lazada = lazada[cols]

all_data = pd.concat([shopee, lazada])
all_data

```
